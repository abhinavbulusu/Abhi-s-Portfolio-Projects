Importing Librarires

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing Dataset

```python
dataset = pd.read_csv('heart.csv')
dataset.head()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

Data Preproccesing

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le6 = LabelEncoder()
le8 = LabelEncoder()
le10 = LabelEncoder()
X[:, 1] = le1.fit_transform(X[:, 1])
X[:, 2] = le2.fit_transform(X[:, 2])
X[:, 6] = le6.fit_transform(X[:, 6])
X[:, 8] = le8.fit_transform(X[:, 8])
X[:, 10] = le10.fit_transform(X[:, 10])


```

Splitting into Test and Training Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Code to test all the Algorithms at the same time using a for loop

```python
# Import necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Define the models to test
models = {
    'Logistic Regression': LogisticRegression(),
    'SVM': SVC(kernel='linear'),
    'KNeighbors': KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(n_estimators=100)
}

# Dictionary to store results
results = {}

# Loop through models, train, and evaluate each
for model_name, model in models.items():
    model.fit(X_train, y_train)  # Train the model
    y_pred = model.predict(X_test)  # Predict on test data
    accuracy = accuracy_score(y_test, y_pred)  # Calculate accuracy
    results[model_name] = accuracy  # Store accuracy in results

# Print results
for model_name, accuracy in results.items():
    print(f"{model_name}: {accuracy:.4f}")
```

Didn't Make sense for Descion Tree and Random Forest to become 1: Need to check for overfitting, using contraints and cross-validation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Decision Tree with constraints
dtree = DecisionTreeClassifier(max_depth=5, max_leaf_nodes=10)
dtree.fit(X_train, y_train)
print("Decision Tree Accuracy:", dtree.score(X_test, y_test))

# Random Forest with constraints
rf = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=0)
rf.fit(X_train, y_train)
print("Random Forest Accuracy:", rf.score(X_test, y_test))

# Cross-validation
scores_dtree = cross_val_score(dtree, X, y, cv=5)
scores_rf = cross_val_score(rf, X, y, cv=5)
print("Cross-Validation (Decision Tree):", scores_dtree.mean())
print("Cross-Validation (Random Forest):", scores_rf.mean())
```

```python
plt.bar(results.keys(), results.values())
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.show()
```
